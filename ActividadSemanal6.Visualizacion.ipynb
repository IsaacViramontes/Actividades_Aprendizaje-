Se puede empezar con el proceso de PCA.

```python
from numpy.ma.core import cumsum
from sklearn.decomposition import PCA

pca = PCA()
pca_total = pca.fit_transform(df_scaled)
pca_summ = pd.DataFrame({"% Varianza Explicada": (np.round(pca.explained_variance_ratio_,4))*100,"% Varianza Acumulada":np.cumsum(pca.explained_variance_ratio_)})

pca_summ.index = ['PC{}'.format(i) for i in range(1, len(pca_summ.index) + 1)]

pca_summ
```

Con la tabla de arriba, podemos observar que con 11 componentes se alcanza una Varianza arriba del 90%.

```python
dfComponentes = pd.DataFrame(pca.components_.round(4), index = df_scaled.columns, columns = pca_summ.index)
dfComponentes.iloc[:,:11]
```

**¿Cuándo se recomienda realizar un PCA y qué beneficios ofrece para Machine Learning?**

PCA (Principal Components Analysis) es una tecnica para atributos continuos. Se utiliza para encontrar los componentes principales del sistema, los cuales son combinaciones lineales de las variables originales y tratan de capturar la maxica varianza posible. Es un procedimiento estadistico que convierte variables correlacionadas en componentes principales linealmente no correlacionadas
